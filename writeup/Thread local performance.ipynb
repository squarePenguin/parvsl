```python
import numpy as np
import matplotlib.pyplot as plt

plt.set_cmap('Set3')
plt.rcParams.update({'font.size': 12})
```

```python
platforms = ['x86 Linux', 'ARMv8 Linux', 'x86 MacOS', 'x86 Win (Visual C++)', 'x86 Win (Cygwin)']
n = 5
ntls = np.array([1.635, 1.605, 1.810, 2.547, 1.985])
tls = np.array([1.655, 4.387, 2.985, 2.653, 51.234])

tls = tls / ntls
```

```python
# create plot
fig, ax = plt.subplots(figsize=(19,7))
index = np.arange(n)
bar_width = 0.35
 
bars = ax.barh(index, tls, bar_width, color="seagreen")
 
ax.set_xlabel('Time slowdown (relative)', size=14)
ax.set_ylabel('Platform', size=14)
ax.set_title('Thread-local performance hit', weight='bold', size=14)
ax.set_yticks(index)
ax.set_yticklabels(platforms)
# ax.set_rasterized(True)
ax.invert_yaxis()

title = ax.title
title.set_weight('bold')
title.set_size(14)

rects = ax.patches

for rect, tl in zip(rects, np.round(tls, decimals=3)): 
    height = rect.get_height()
    width = rect.get_width()
    ax.text(width + 0.1, rect.get_y() + height / 2, tl, ha='left', va='center')

# plt.tight_layout()
```
